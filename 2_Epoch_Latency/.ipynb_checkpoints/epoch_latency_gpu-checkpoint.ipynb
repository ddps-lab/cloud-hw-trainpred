```python
import numpy as np # linear algebra
import pandas as pd 
import pickle
import os
import glob
import datetime
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import LabelEncoder
pd.options.display.float_format = '{:.10f}'.format
from matplotlib import rcParams
%matplotlib inline
rcParams['figure.figsize'] = 14,6
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
from luminol import anomaly_detector
from luminol.anomaly_detector import AnomalyDetector
from luminol.correlator import Correlator
from scipy import stats
from matplotlib.backends.backend_pdf import PdfPages
from cydets.algorithm import detect_cycles
import scipy.fft
from datetime import datetime
import pandas as pd
import io
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import math
import jenkspy
from sklearn.cluster import MeanShift, estimate_bandwidth
from scipy.signal import savgol_filter
from scipy.signal import find_peaks, peak_prominences
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_percentage_error, r2_score, mean_squared_error



all_true_df = []  # 원조 1 epoch latency 평균
all_pred_df_gpu = []

def find_nearest(array, value): # 제일근처값찾기
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]

def scoreLuminolALLData(ts):   
    ts_dict = ts.to_dict()
    score_v = []
    my_detector = AnomalyDetector(ts_dict)
    score = my_detector.get_all_scores()
    for timestamp, value in score.iteritems():
        #print(timestamp, value)
        score_v.append(value)
    return score_v

def retFig(x, y): # 결과 pdf로 저장하기용도
    fig = plt.figure()
    a= plt.plot(x, y)
    return fig

def min_max_normalize(lst):
    normalized = []
    
    for value in lst:
        normalized_num = (value - min(lst)) / (max(lst) - min(lst))
        normalized.append(normalized_num)
    return normalized
```

#### Data.csv 전처리

```python
# Data.csv 데이터 전처리
Data_T = pd.read_csv(path_list[0]+'/Data.csv')

# 전처리 ~~~~---------------------------------------------------------------------
Data_new_timestamp = pd.DataFrame( Data_T['timestamp'] )
Data_new_timestamp = Data_new_timestamp.replace(r' ', '', regex=True)
Data_new_timestamp = Data_new_timestamp.replace(r':', '', regex=True)
Data_new_timestamp = Data_new_timestamp.replace(r'/', '', regex=True)
Data_new_timestamp = Data_new_timestamp.astype('float')

Data_T = Data_T.replace(r'W', '', regex=True)
Data_T = Data_T.replace(r' ', '', regex=True)
Data_T = Data_T.replace(r'MHz', '', regex=True)
Data_T = Data_T.replace(r'MiB', '', regex=True)
Data_T = Data_T.replace(r'%', '', regex=True)
not_float_feature=[ ' clocks.current.sm [MHz]',' utilization.gpu [%]',' utilization.memory [%]',' power.draw [W]']     
for i in not_float_feature:
    Data_T[i]=Data_T[i].astype('float')

Data_T.rename(columns = {' temperature.gpu' : 'temperature.gpu'}, inplace = True)
Data_T.rename(columns = {' utilization.memory [%]' : 'utilization.memory[%]'}, inplace = True)
Data_T.rename(columns = {' utilization.gpu [%]' : 'utilization.gpu[%]'}, inplace = True)
Data_T.rename(columns = {' power.draw [W]' : 'power.draw[W]'}, inplace = True)
Data_T.rename(columns = {' clocks.current.sm [MHz]' : 'clocks.current.sm[MHz]'}, inplace = True)
```

```python
path_model_list = ['MNIST_CNN*','AlexNet*','CIFAR10*','InceptionV3*',
                     'VGG19*','ResNet50*','InceptionResNetV2*','LeNet5*',
                   'ResNetSmall*'] 
path_model_list
```

```python
T_all_true_df = []  # total  
T_all_pred_df_gpucpu =[] # total 

for path in path_list:
    for path_model in path_model_list:
        all_true_df = [] # 모델별로 
        all_pred_df_gpucpu =[]  # 모델별로 
        try:  # 해당모델이 없으면 에러무시하고 넘어가기
            filenames = glob.glob(str(path)+'/'+str(path_model) )
            print(filenames)  # 해당 모델의 머선 파일들이 있는지
            
            for file in filenames: # 해당 모델이름으로 필터링된 파일 모두살펴보기
                print(file) #이 파일위주로 볼것이다
                 
                # 1) csv 파일 (모델, 배치사이즈등등별로 다름) ----------------------------------------------------------------------------------------
                with open(file, 'rb') as fr:
                    epoch_start_end_timestamp = pickle.load(fr)

                epoch_start_timestamp_array = []  # epoch 시작위치
                epoch_end_timestamp_array = []   # epoch 끝위치
                epoch_start_end_timestamp_Dataframe = pd.DataFrame(epoch_start_end_timestamp)

                for i in range(len(epoch_start_end_timestamp.keys())):
                    epoch_start_end_timestamp_Dataframe = epoch_start_end_timestamp_Dataframe.replace(r' ', '', regex=True)
                    epoch_start_end_timestamp_Dataframe = epoch_start_end_timestamp_Dataframe.replace(r':', '', regex=True)
                    epoch_start_end_timestamp_Dataframe = epoch_start_end_timestamp_Dataframe.replace(r'-', '', regex=True)
                    epoch_start_end_timestamp_Dataframe = epoch_start_end_timestamp_Dataframe.replace(r'/', '', regex=True)
                for i in range(len(epoch_start_end_timestamp_Dataframe.columns)):
                    epoch_start_timestamp_array.append(epoch_start_end_timestamp_Dataframe[i][0])
                    # 시작지점 리스트로
                for i in range(len(epoch_start_end_timestamp_Dataframe.columns)):
                    epoch_end_timestamp_array.append(epoch_start_end_timestamp_Dataframe[i][1])
                    # 끝지점 리스트로
                    
                print( '[총 에포크수]  : ' + str(  len(epoch_start_end_timestamp.keys())))
                
                # 2) Data.csv (Data_T) -----------------------------------------------------------------------------------------
                #Data_T['epoch'] = 0 
                
                # epoch 시작 끝지점이 Data_T 랑 완전히 같지않기 때문에 가장 가까운 지점만 골라내줌 -------------------------------------
                new_timestamp_array = np.array(Data_new_timestamp['timestamp'])
                epoch_start_timestamp_array_nearest = []
                epoch_end_timestamp_array_nearest = []

                for i in epoch_start_timestamp_array : 
                    what = (find_nearest (new_timestamp_array, float(i)) )
                    epoch_start_timestamp_array_nearest.append(float(what))
                for i in epoch_end_timestamp_array : 
                    what = (find_nearest (new_timestamp_array, float(i)) )
                    epoch_end_timestamp_array_nearest.append(float(what))

                # epoch 별로 끊어지는 지점 index 찾아내기 ---------------------------------------------------
                epoch_start_list = []  # 시작지점 index
                epoch_end_list = []  # 끝지점 index
                for i in range(len(epoch_start_timestamp_array_nearest)):
                    epoch_start_list.append(Data_new_timestamp [Data_new_timestamp["timestamp"] == epoch_start_timestamp_array_nearest[i]].index[0])
                for i in range(len(epoch_end_timestamp_array_nearest)):
                    epoch_end_list.append(Data_new_timestamp [Data_new_timestamp["timestamp"] == epoch_end_timestamp_array_nearest[i]].index[0])

                # Data_T2  =  해당 모델 데이터만 가지고 살펴본다 ------------------------------------------------------------------------
                # 시작한지점이랑 끝난지점 기준으로 데이터 앞뒤 잘라내줌.-------------------------------------------------------
                
                Data_T2 = Data_T[epoch_start_list[0]:epoch_end_list[len(epoch_start_end_timestamp.keys())-1]]
                Data_T2 = Data_T2.reset_index()
                
                if(len(Data_T2)>300 and len(Data_T2)<1999*10):   # 총길이가 100미만인거는 결과가 너무안좋아서 뺌
                    # 잘라주고 나서 epoch start, end 리스트 재배치 ------------------------------------------
                    start = epoch_start_list[0] 
                    for i in range(len(epoch_start_list)):
                        epoch_start_list[i] = epoch_start_list[i] - start
                    for i in range(len(epoch_end_list)):
                        epoch_end_list[i] = epoch_end_list[i] - start
                    
                    # epoch 매겨주기---------------------------------------------------------------------
                    epoch_num = 0
                    Data_T2['epoch'] = 0 
                    for i in range(len(Data_T2)):
                        if i in epoch_start_list:
                            epoch_num = epoch_num + 1
                        Data_T2.iloc[i:i+1]['epoch'] = epoch_num
                    
                    for i in range(len(Data_T2)):
                        Data_T2['timestamp'][i] = Data_T2['timestamp'][i][10:21]
                    Data_T2['date'] = pd.to_datetime(Data_T2['timestamp'], format='%H:%M:%S.%f')
                    
                    #dataframe 이 2/22 2:2:2.000 인경우 뒤의 0을 제외하고 [:-3 으로 계산해서 에러나기때문]
                    if(len(str(Data_T2['date'][len(Data_T2)-1])[:-3]) != 26):
                        d2 = datetime.strptime(str(Data_T2['date'][len(Data_T2)-2])[:-3], "%Y-%m-%d %H:%M:%S.%f")
                    else:
                        d2 = datetime.strptime(str(Data_T2['date'][len(Data_T2)-1])[:-3], "%Y-%m-%d %H:%M:%S.%f")

                    if(len(str(Data_T2['date'][0])[:-3]) != 26):
                        d1 = datetime.strptime(str(Data_T2['date'][1])[:-3], "%Y-%m-%d %H:%M:%S.%f")
                    else:
                        d1 = datetime.strptime(str(Data_T2['date'][0])[:-3], "%Y-%m-%d %H:%M:%S.%f")
                        
                    print(  '[ 1 EPOCH latency 평균 ? ]  :  ' + str((d2-d1)/10))
                    total_latency_millisecond = (d2-d1).total_seconds()*1000
                    total_latency_second = (d2-d1).total_seconds()
                    one_epoch_average_millisecond = total_latency_millisecond/10
                    one_epoch_average_second = total_latency_second/10
                    #print( '[ 1 EPOCH latency 평균 ?(millisecond 단위) ]  :' + str(total_latency_second) )
                    all_true_df.append(one_epoch_average_second)
                    T_all_true_df.append(one_epoch_average_second)
                    
                
                    # -------------------------------------------------------------------------------------
                    # ----1999 (루미놀 적용)-----------------------------------------------------------------------------
                    Data_T2 = Data_T2[int(len(Data_T2)/2):]
                    Data_T2 = Data_T2[0:1999] # 루미놀의 한계
                    start_index = Data_T2[0:1].index.start
                    for i in range(len(epoch_start_list)):
                        epoch_start_list[i] = epoch_start_list[i]-start_index
                    newList = [x for x in epoch_start_list if x>0]

                    Data_T2 = Data_T2.reset_index()
                    
                    if(len(Data_T2) < 800):
                        holsu = 21
                        if(len(Data_T2) < 200):
                            holsu = 11
                    else:
                        holsu = 91
                    if(len(Data_T2) > 1500):
                        holsu = 101
                    if(len(Data_T2) > 1900):
                        holsu = 231
                        
                    # 루미놀, 필터 1th --------------------------------------
                    Data_T2['sum'] = Data_T2['utilization.gpu[%]']   
                    Data_T2['luminol_first'] = scoreLuminolALLData(Data_T2['sum'])              
                    data = np.array(Data_T2['luminol_first'] )
                    w = savgol_filter(data,holsu, 1)                    
                    wseries = pd.Series(w)
                    # 루미놀결과값에 normalization 해준다음, 필터 2th + Normalization 해줌
                    Data_T2['luminol_first_normalize'] = min_max_normalize(wseries)
                    data = np.array(Data_T2['luminol_first_normalize'] )
                    w = savgol_filter(data, holsu , 1)
                    wseries = pd.Series(w)
                    Data_T2['luminol'] = min_max_normalize(wseries)
                    
                    # 피크값 뽑아내기 -------------------------------------------------------
                    graph_data = Data_T2['luminol']
                    data = list(graph_data)
                    peaks, _ = find_peaks(data, prominence =0.01) 
                    df_high_score_ = list(peaks)
                
                    for i in df_high_score_:
                        if(data[i] < Data_T2['luminol'].mean()):
                            df_high_score_.remove(i)
                        
                    #----------------------------------------------------------------------
                    anomaly_pd = pd.DataFrame(df_high_score_)
                    anomaly_pd['index'] = anomaly_pd.index
                    anomaly_pd.columns = ['x', 'y']
                    a = anomaly_pd['x']
                    b = 0 # 1차원으로 만들어주기위해
                    eps  = 0.5
                    
                    anomaly_pd_list = list(anomaly_pd['x'])
                    anomaly_pd['y'] = 1
                    
                    X = anomaly_pd.to_numpy()
                    dbscan =  DBSCAN(eps , min_samples=1)
                    clusters = dbscan.fit_predict(X)

                    if(len(str(Data_T2['date'][df_high_score_[0]])) != 26):
                        d2_1999 = datetime.strptime(str(Data_T2['date'][df_high_score_[0]+1])[:-3], "%Y-%m-%d %H:%M:%S.%f")
                    else:
                        d2_1999 = datetime.strptime(str(Data_T2['date'][df_high_score_[0]])[:-3], "%Y-%m-%d %H:%M:%S.%f")
                    if(len(str(Data_T2['date'][df_high_score_[-1]])) != 26):
                        d1_1999 = datetime.strptime(str(Data_T2['date'][df_high_score_[-1]-1])[:-3], "%Y-%m-%d %H:%M:%S.%f")
                    else:
                        d1_1999 = datetime.strptime(str(Data_T2['date'][df_high_score_[-1]])[:-3], "%Y-%m-%d %H:%M:%S.%f")

                    total_latency_millisecond = (d1_1999-d2_1999).total_seconds()*1000
                    total_latency_second2 = (d1_1999-d2_1999).total_seconds()
                    
                    cluster_num = int(max(clusters))
                    if(cluster_num == 0 ):  # 클러스터가 0이면 나눌때 에러가 나기때문.
                        cluster_num = 1
                        predict_epoch_num = total_latency_second2
                    else:
                        predict_epoch_num = total_latency_second2/cluster_num
            
                    if(total_latency_second2 == 0):
                        predict_epoch_num = total_latency_second2                     
                
                    all_pred_df_gpucpu.append(predict_epoch_num)   
                    T_all_pred_df_gpucpu.append(predict_epoch_num)   
                    
                    print('[ 예측한 1 EPOCH latency 평균 ? ]  :  ' + str(predict_epoch_num)) 
                    print("클러스터 레이블 : \n{}".format(clusters))
                    print("클러스터 갯수 : " + str(cluster_num))
                    print("sum 데이터 분산  :  " + str(Data_T2['utilization.gpu[%]'].var()))
                    print("luminol 분산  :  " + str(Data_T2['luminol'].var()))
                    pjpc = math.sqrt(Data_T2['luminol'].var()) 
                    print("luminol 표준편차  :  " + str(pjpc))
                    print("luminol 평균값 : " + str(Data_T2['luminol'].mean()))
                    print("luminol 중앙값 : " + str(Data_T2['luminol'].median()))

                    # 결과 그래프 Pdf 로 저장하기  --------------------------------------------------------------------------------------------------------
                    filename = file
                    with PdfPages(str(path)+'_'+str(filename[int(filename.find('/'))+1:-4]) + '.pdf') as pdf:
                        print(str(path)+'_'+str(filename[int(filename.find('/'))+1:-4]) + '.pdf')

                        plt.plot(Data_T2['sum'])
                        plt.show()
                        plt.plot(range(len(data)), data)
                        plt.scatter(df_high_score_, [data[t] for t in df_high_score_], c='red')
                        for i in newList:
                            if(i < 1999):
                                plt.axvline(x=[i], color='r', linestyle=':', linewidth=1)
                        plt.show()
                        
        
            print("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@")
            print(path_model)
            
            all_pred_df = all_pred_df_gpucpu
            print("mae")
            print(mean_absolute_error(all_true_df, all_pred_df))

            print("rsme")
            MSE = mean_squared_error(all_true_df, all_pred_df) 
            print(np.sqrt(MSE))

            print('mape')
            print(mean_absolute_percentage_error(all_true_df, all_pred_df)*100)

            print('r2')
            print(r2_score(all_true_df, all_pred_df))
            
        except:
            pass
```

```python
all_pred_df = T_all_pred_df_gpucpu
all_true_df = T_all_true_df

from sklearn.metrics import mean_absolute_error
print("mae")
print(mean_absolute_error(all_true_df, all_pred_df))

from sklearn.metrics import mean_squared_error 
print("rsme")
MSE = mean_squared_error(all_true_df, all_pred_df) 
print(np.sqrt(MSE))

from sklearn.metrics import mean_absolute_percentage_error, r2_score, mean_squared_error
print('mape') 
print(mean_absolute_percentage_error(all_true_df, all_pred_df)*100)

print('r2')
print(r2_score(all_true_df, all_pred_df))
```
