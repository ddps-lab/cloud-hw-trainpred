```python
path_list = ['g5','g4dn','p3','p100'] 
```

```python
import numpy as np # linear algebra
import pandas as pd 
import pickle
```

```python
fn = r'./data/'+path_list[0]+'/devicequery.csv'  # 위치조심
data=pd.read_csv(fn,header=None)
data.columns = ['features',path_list[0]]
total_data = data
```

```python
for i in path_list[1:]:
    fn = r'./data/'+str(i)+'/devicequery.csv'  # 위치조심
    data=pd.read_csv(fn,header=None)
    data.columns = ['features',str(i)]
    total_data = pd.concat([total_data, data[str(i)]], axis=1)
```

```python
len(total_data)
total_data['g5'][2]
```

```python
total_data

for i in range(len(total_data)):
    if((total_data['g5'][i] == total_data['g4dn'][i]) & (total_data['g5'][i] == total_data['p3'][i])):  
        total_data = total_data.drop(i, axis=0)
total_data = total_data.reset_index(drop=True)
total_data = total_data.set_index('features',drop=False)
total_data
```

```python
# 초반에 데이터 수집할때 Memory Bus Width/1000로 수집해서 후처리해주는거임
for i in path_list:
    total_data[i]['Memory Bus Width'] = float(total_data[i]['Memory Bus Width'])*1000


data = {
    'features': 'Memory Bandwidth',
    'g5' : str( float(total_data['g5']['Memory Clock rate']) * float(total_data['g5']['Memory Bus Width']) ),      
    'g4dn' : str( float(total_data['g4dn']['Memory Clock rate']) * float(total_data['g4dn']['Memory Bus Width']) ),      
    'p3' : str( float(total_data['p3']['Memory Clock rate']) * float(total_data['p3']['Memory Bus Width']) ),  
    'p100' : str( float(total_data['p3']['Memory Clock rate']) * float(total_data['p3']['Memory Bus Width']) ),      
}
total_data = total_data.append(data, ignore_index=True)
total_data = total_data.set_index('features',drop=True)
total_data
```

```python
total_data
```

```python
total_data.to_csv('devicequery_result.csv')
```
